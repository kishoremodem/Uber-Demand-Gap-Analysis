```python
d['Request timestamp']=pd.to_datetime(d['Request timestamp'])               
#to convert the Request timestamp into proper time and date formate 
```

```python
d['Drop timestamp']=pd.to_datetime(d['Drop timestamp'])           
#to convert the Drop timestamp into proper time and date formate 
```

```python
d.head()                            #After converting the request timestamp and drop timestamp into proper formate
```

```python
d['Request Date'] = [i.date() for i in d['Request timestamp']] 
#to seperate the date and time into seperate columns from request timestamp 
```

```python
d['Drop Date'] = [i.date() for i in d['Drop timestamp']]              
#to seperate the date and time into seperate columns from Drop timestamp 
```

```python
d['Request Time'] = [i.time() for i in d['Request timestamp']]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
d['Drop Time'] = d['Drop timestamp'].dt.time
```

```python
d['day'] = d['Request timestamp'].dt.day_name()         #creating new column which contain day of the date
```

```python
plt.title('Status_Ride' )       #the graph tells how many trip are complated and how many are cancelled and how many are 
                               #no cars are availble
sns.countplot(d['Status'])
```

```python
p = plt.subplots(figsize=(30,20))                 #trip completed ,cancelled and no car Avaliable ratio in various days of the week
plt.title('days_status',fontweight="bold")
p = sns.countplot(x='day', hue='Status',data=d)   #in wednesday more request for cabs with campare to remining days
```

```python
d['hours'] = d['Request Time'].apply(lambda x:x.hour)                #creating a new column which contain hours
```

```python
d.head()
```

```python
q = plt.subplots(figsize=(20,10))                              # request for cabs at various hours from the graph  
                                                               # 4 to 10 and 17to 23 high request for cabs
plt.title('Hours',fontweight="bold")
q = sns.countplot(x='hours',data=d)
```

```python
z = plt.subplots(figsize=(20,10))                       # more request for city to airport at morning time and more request
                                                        # from airport to city at evening time
plt.title('locations')
z = sns.countplot(x='hours',hue='Pickup point', data=d)
```

```python
city=d[d['Pickup point']=='City']
airport=d[d['Pickup point']=='Airport']
```

```python
p = plt.subplots(figsize=(30,20))
plt.title('airport to city')                                #cars are not avalible from airport city at evening time where 
                                                            #there is requiment
p = sns.countplot(x='hours', hue='Status',data=airport)
```
